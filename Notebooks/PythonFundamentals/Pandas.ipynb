```python
# pandas é uma biblioteca para tratar dados tabulares, uma estrutura comum em ciencias de dados
# tem as características de um banco onde temos diferentes colunas com diferentes tipos de dados
import pandas as pd
```

```python
# carrega arquivo para o dataframe pandas
dados = pd.read_csv('Credit.csv')

# ver o formato
dados.shape
```

```python
# resumo estatístico de colunas numéricas
dados.describe()
```

```python
# mostra os primeiros registros da tabela
dados.head()
```

```python
# mostra os últimos registros com parametros
dados.tail(2)
```

```python
# filtros por nome de coluna
dados[["duration"]]
```

```python
# filtros de linhas por índice
dados.loc[1:3]
```

```python
# filtros por lista de valores das colunas específicas
dados.loc[[1, 3]]
```

```python
# filtro passando parametros da coluna e valor
dados.loc[dados['purpose'] == 'radio/tv']
```

```python
# filtro passando parametros de outra coluna e valor
dados.loc[dados['credit_amount'] > 18000]
```

```python
# atribuindo resultado a outra variável criando outro data frame
credito2 = dados.loc[dados['credit_amount'] > 18000]
print(credito2)
```

```python
# definindo só algumas colunas
credito3 = dados[['checking_status', 'duration']].loc[dados['credit_amount'] > 18000]
print(credito3)
```

```python
# criando uma série com uma única coluna
# pode ser criada a partir de listas, array do numpy ou coluna do dataframe

# series a partir de uma lista lista
s1 = pd.Series([2, 5, 3, 34, 54, 23, 1, 16])
print(s1)
type(s1)
```

```python
# series a partir do array do numpy
import numpy as np
array1 = np.array([2, 5, 3, 34, 54, 23, 1, 16])
s2 = pd.Series(array1)
print(s2)
type(s2)
```

```python
# series a partir do data frame
s3 = dados['purpose']
print(s3)
type(s3)
```

```python
# com dois colchetes faz virar um dataframe
s4 = dados[['purpose']]
print(s4)
type(s4)
```

```python
# renomear objetos do pandas
dados.rename(columns={'duration':'duração', 'purpose':'propósito'})
```

```python
# porém a alteração não fica persistida e o que foi renomeado volta
dados.head(1)
```

```python
# para persistir precisamos usar o inplace=True
dados.rename(columns={'duration':'duração', 'purpose':'propósito'}, inplace=True)
dados.head(1)
```

```python
# excluir coluna
dados.drop('checking_status', axis=1, inplace=True)
print(dados)
```

```python
# verificar como ficou o drop
dados.head(1)
```

```python
# verificar dados nulos
# dados faltantes são um problema cronico em ciencia de dados
# existem métodos em ML que não comportam esse tipo de dados e precisam ser tratados
dados.isnull()
```

```python
# visualizar dados nulos com o sum é melhor
dados.isnull().sum()
```

```python
# remover dados com 'na'
dados.dropna()
```

```python
# preencher dados faltantes
dados['duração'].fillna(0, inplace=True)
```

```python
# iloc
dados.iloc[0:3, 0:5] # linhas de 0 a 3 e colunas de 0 a 5
```

```python
# iloc
dados.iloc[[0, 1, 2, 3, 7], 0: 5] # linhas 0, 1, 2, 3 e 7 especificamente e colunas de 0 a 5
```

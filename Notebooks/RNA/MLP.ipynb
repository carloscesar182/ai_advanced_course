```python
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# carregar os dados de ires e dividir em treino e teste
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)
```

```python
# criar o modelo mpl
modelo = MLPClassifier(
    verbose=True, # pra acompanhar de forma iterativa
    hidden_layer_sizes=(5,4) , # define o tamanho da camada oculta e neuronios
    activation='relu', # um dos elementos principais que faz a rede neural funcionar
    batch_size=20,
    learning_rate='adaptive', # não formamos um valor mas sim como a taxa vai ser atualizada
    momentum=0.9,
    early_stopping=False, # define se para o treinamento antecipadamente se não ver melhoria
    max_iter=1000, # quantas epochs max
    random_state=10
  )
modelo.fit(X_train, y_train)
```

```python
# gerar um grafico que mostra a melhoria da loss funcion
plt.plot(modelo.loss_curve_)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
# fazer as previsões
previsoes = modelo.predict(X_test)
previsoes
```

```python
# medir a acuracia
print(accuracy_score(y_test, previsoes))
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# importar arquivo
# utilizar line strip pra dividir os elementos que estão com virgula
with open('Transacoes.txt', "r") as f:
  transactions = [line.strip().split(',') for line in f.readlines()]
transactions
```

```python
# criar uma lista onde cada elemento da lista tenha uma lista com as transações
# nesse formato, conseguimos usar facilmente o TransactionEncoder pra transportar os elementos pra colunas
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)
df
```

```python
# minerar os itens frequentes usando apriori
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
# criar as regras definindo a confiança mínima
# usar o módulo associationrules pra mostrar as regras que o apriori encontra
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
import pygad
import numpy as np
```

```python
# equação 2 * x + 5 = 20
def equation_func(ga_instance, solution, solution_idx): # parametros que a função coordena
  x = solution[0]
  equation_output = 2 * x + 5
  target_output = 20
  fitness = 1.0 / np.abs(equation_output - target_output + 1e-8)
  return fitness

# criar uma instancia do algoritmo genético
ga_instance = pygad.GA(
    num_parents_mating=10,
    fitness_func=equation_func, # função que criamos
    sol_per_pop=20, # população
    num_genes=1, # número de genes
    gene_type=float,
    init_range_low=-10,
    init_range_high=10,
    num_generations=100
)

ga_instance.run()

# grafico da convergencia da solução
ga_instance.plot_fitness()

solution, solution_fitness, _ = ga_instance.best_solution()
print("Fitness da melhor solução: ", round(solution[0], 1))
print("Pontos: ", solution_fitness)
```

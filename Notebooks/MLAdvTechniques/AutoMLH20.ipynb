```python
import h2o
from h2o.automl import H2OAutoML
```

```python
# iniciar o cluster de forma default
h2o.init()
```

```python
# importar o arquivo
data = h2o.import_file('insurance.csv')
```

```python
# visualizar o arquivo
data
```

```python
# dropar a coluna 1 que não serve pra nada
data = data.drop("C1")
```

```python
# dividir treino e teste
train, test = data.split_frame(ratios=[0.7], seed=0)
```

```python
# rodar nossa função de automl
model_automl = H2OAutoML(max_runtime_secs=180, sort_metric='AUTO') # metrica da área sobre a curva
# vai processar por 180 segundos e trazer várias metricas do melhor modelo
model_automl.train(y='Accident', training_frame=train)
```

```python
# ranking dos modelos que foi utilizado
ranking = model_automl.leaderboard
ranking = ranking.as_data_frame()
ranking
```

```python
# fazer as previsões
pred = model_automl.leader.predict(test)
pred
```
